### Mapping GMP

```python
import pandas as pd
```

```python
# CSV file contains city names, their BLS codes (Metropolitan 
# MTAs/NECTAs only), and their latitude and longitude
MSA = pd.read_csv('MSA.csv')

# BEA data on Real Gross Metropolitan Product
data = pd.read_csv('allgmp.csv', index_col='GeoFIPS', skipfooter=3, engine='python')[87:]
```

```python
col_list = ['GeoName'] + [str(i) for i in list(range(2001, 2017))]
df = data[(data['IndustryId'] == '1') & (data['ComponentId'] == '900')][col_list]
df['change'] = df['2016'].astype(float) / df['2001'].astype(float)
df['change'] = (df['change'] - 1) * 100
#df.sort_values('change').T
```

```python
df[df['change'] < -1].sort_values('change').T
```
